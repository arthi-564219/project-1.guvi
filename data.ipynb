```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()  # Make sure chromedriver is installed and in PATH in 2024 movie list
#IMDB URL for movie released data in 2024
#url="https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31"
url = "https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=adventure,animation"
#open the URL
driver.get(url)

#allow the page to load (you can adjust the sleep time if necessary)
time.sleep(3)

#initialize lists to store scraped the data
titles = []
ratings = []
votings = []
durations = []
#loop through all movie item on the page
#//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul
#//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul
movie_items = driver.find_elements(By.XPATH,'//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li')#//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul

for movie_item in movie_items:#//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[1]  
    try:#//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[1]   / /div/div/div/div[1]/div[2]/div[1]/a/h3
        #Extract the movie title
        title = movie_item.find_element(By.XPATH,'. /div/div/div/div[1]/div[2]/div[1]/a/h3').text

        #Extract the movie rating
        rating = movie_item.find_element(By.XPATH,'./div/div/div/div[1]/div[2]/span/div/span/span[1]').text

        #Extract the number of votes
        voting = movie_item.find_element(By.XPATH,'./div/div/div/div[1]/div[2]/span/div/span/span[2]').text

        #Extract the movie duration//#//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[1]/div/div/div/div[1]/div[2]/div[2]/span[2]
        duration = movie_item.find_element(By.XPATH,'./div/div/div/div[1]/div[2]/div[2]/span[2]').text

        #Append the data to the lists
        titles.append(title)
        ratings.append(rating)
        votings.append(voting)
        durations.append(duration)

    except Exception as e:
        #if any were element is not founf of,skip the data this movie and print the error occuracy message shown
        print(f"Error extracting data for a movie: {e}")  
        continue

#creating a dataframe using data the extracted 
df = pd.DataFrame({
    'Title':titles,
    'Rating':ratings,
    'Votes':votings,
    'Duration':durations
})


#save the dataframe to csv file(optional) this data
##extract data loading from next step on the exception or creating dataframe 
#df.to_csv('imdb_movie_2024.csv',index=false)

#close the this web brower 
driver.quit()
df
```
