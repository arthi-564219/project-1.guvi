```python
import pandas as pd

# Example data
data = {'Name': ['Alice', 'Bob', 'Alice'],
        'Age': ['25', '30', '25']}
df = pd.DataFrame(data)

# View columns
print(df.columns)

# Convert 'Age' column to int
df['Age'] = df['Age'].astype(int)

# Drop duplicate rows
df.drop_duplicates(inplace=True)

print(df)
```
